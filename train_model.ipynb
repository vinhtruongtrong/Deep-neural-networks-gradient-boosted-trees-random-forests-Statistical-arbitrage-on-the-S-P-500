```python
import import_ipynb
import pre_processing as pp
```

```python
data = pp.get_data()
```

Get data after preprocessing and sort by column date and name.

```python
data.sort_values(by=['date','Name'])
```

```python
X = data
y = data.label
```

Define features and label.

```python
from sklearn.model_selection import train_test_split
_X_train, _X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle = True)
X_train = _X_train.copy().drop(['date','Name','price','cross_sectional_median'],axis = 1)
X_test = _X_test.copy().drop(['date','Name','price','cross_sectional_median'],axis = 1)
```

Split into train dataset and test dataset.

```python
from sklearn import metrics
from sklearn import ensemble
from sklearn.neural_network import MLPRegressor
models = [
    ensemble.RandomForestRegressor(n_estimators=100, random_state=0),
    ensemble.GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0, loss='ls'),
    MLPRegressor(solver='adam',
                  alpha=0.0001,
                  activation='relu',
                    batch_size=150,
                    hidden_layer_sizes=(200, 100),
                    random_state=1)
]
```

Define 3 model : Random Forest, Gradient Boosting Decision Tree, DNN.

```python
for item in models:
    regressor = item
    regressor.fit(X_train, y_train)  
    y_pred = regressor.predict(X_test)
    _X_test[item.__class__.__name__] = y_pred
```

```python
_X_test['ENS'] = _X_test[['RandomForestRegressor','GradientBoostingRegressor','MLPRegressor']].mean(axis = 1)
```

Compute value of column ENS = median value of RF, GBT, DNN.

```python
_X_test.sort_values(by = ['RandomForestRegressor','GradientBoostingRegressor','MLPRegressor','ENS'],ascending = False)
```
