```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
```

Extract price of stock to 31 features.

```python
def extract_feature(f0Data):
    f0Data['f1'] = f0Data.price / f0Data.price.shift(1) - 1
    f0Data['f2'] = f0Data.price / f0Data.price.shift(2) - 1
    f0Data['f3'] = f0Data.price / f0Data.price.shift(3) - 1
    f0Data['f4'] = f0Data.price / f0Data.price.shift(4) - 1
    f0Data['f5'] = f0Data.price / f0Data.price.shift(5) - 1
    f0Data['f6'] = f0Data.price / f0Data.price.shift(6) - 1
    f0Data['f7'] = f0Data.price / f0Data.price.shift(7) - 1
    f0Data['f8'] = f0Data.price / f0Data.price.shift(8) - 1
    f0Data['f9'] = f0Data.price / f0Data.price.shift(9) - 1
    f0Data['f10'] = f0Data.price / f0Data.price.shift(10) - 1
    f0Data['f11'] = f0Data.price / f0Data.price.shift(11) - 1
    f0Data['f12'] = f0Data.price / f0Data.price.shift(12) - 1
    f0Data['f13'] = f0Data.price / f0Data.price.shift(13) - 1
    f0Data['f14'] = f0Data.price / f0Data.price.shift(14) - 1
    f0Data['f15'] = f0Data.price / f0Data.price.shift(15) - 1
    f0Data['f16'] = f0Data.price / f0Data.price.shift(16) - 1
    f0Data['f17'] = f0Data.price / f0Data.price.shift(17) - 1
    f0Data['f18'] = f0Data.price / f0Data.price.shift(18) - 1
    f0Data['f19'] = f0Data.price / f0Data.price.shift(19) - 1
    f0Data['f20'] = f0Data.price / f0Data.price.shift(20) - 1
    f0Data['f21'] = f0Data.price / f0Data.price.shift(40) - 1
    f0Data['f22'] = f0Data.price / f0Data.price.shift(60) - 1
    f0Data['f23'] = f0Data.price / f0Data.price.shift(80) - 1
    f0Data['f24'] = f0Data.price / f0Data.price.shift(100) - 1
    f0Data['f25'] = f0Data.price / f0Data.price.shift(120) - 1
    f0Data['f26'] = f0Data.price / f0Data.price.shift(140) - 1
    f0Data['f27'] = f0Data.price / f0Data.price.shift(160) - 1
    f0Data['f28'] = f0Data.price / f0Data.price.shift(180) - 1
    f0Data['f29'] = f0Data.price / f0Data.price.shift(200) - 1
    f0Data['f30'] = f0Data.price / f0Data.price.shift(220) - 1
    f0Data['f31'] = f0Data.price / f0Data.price.shift(240) - 1
    return f0Data
```

Read stock data without column volume.

```python
dataset = pd.read_csv('all_stocks_5yr.csv')
dataset = dataset.drop(['volume'],axis = 1)
dataset.head()
```

Define some missing stock data (number < 1259) and remove it.

```python
missing_list = ['APTV','DWDP','DWDP','BHF','BHGE','DXC','HLT','EVH','FTV','UA','WLTW','CSRA','HPQ','HPQ','HPE','PYPL','EVHC',
                'KHC','WRK','QRVO','CFG','SYF','INFO','NAVI','GOOG','ALLE','FOX','FOXA','NWS','NWSA','COTY','IQV','O','DHR',
                'ES','ICE','BMY','ORCL']
for ml in missing_list:
    dataset = dataset[dataset.Name != ml]
```

```python
dataset.groupby(['Name']).size().sort_values()
```

Define price by median value of 4 columns: open, close, high, close.

```python
dataset['price'] = dataset[['open','close','high','low']].mean(axis = 1).round(2)
dataset.head()
```

```python
df = dataset.drop(['open','close','high','low'],axis = 1)
df.columns
```

Add 31 features columns with NaN value.

```python
for i in range(31):
    df.insert(loc=i+3,column = str('f'+str(i+1)),value = np.nan)
df.head()
```

Fill value into 31 feature-columns.

```python
final_data = df.groupby(['Name']).apply(extract_feature)
final_data.head()
```

```python
final_data = final_data.dropna()
final_data.head()
```

Drop all of rows have NaN value in any column in 31 feature-columns.

```python
temp = final_data.drop(['Name','price','date'],axis = 1)
temp['label'] = temp.mean(axis = 1)
final_data['cross_sectional_median'] = temp.label
```

```python
final_data.loc[final_data['f1'] > final_data['cross_sectional_median'], 'label'] = 1 
final_data.loc[final_data['f1'] <= final_data['cross_sectional_median'], 'label'] = 0 
```
